```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
costs = pd.read_csv('energy_costs_.csv')
roi = pd.read_csv('roi_analysis_.csv')

# Cost comparison chart
plt.figure(figsize=(12, 6))
categories = costs['Category']
current = costs['Current_Monthly_TRY']
proposed = costs['Proposed_Monthly_TRY']

x = range(len(categories))
plt.bar(x, current, width=0.4, label='Current Model (Fossil Fuel)', color='#d9534f')
plt.bar([i + 0.4 for i in x], proposed, width=0.4, label='SÜGEP Model (PV-Integrated)', color='#5cb85c')

plt.xticks([i + 0.2 for i in x], categories)
plt.ylabel('Monthly Cost (TRY)')
plt.title('Economic Transformation: Monthly Household Expenses')
plt.legend()
plt.savefig('cost_comparison.png')

# Return on investment (ROI) simulation
plt.figure(figsize=(12, 6))
plt.plot(roi['Month'], roi['Cumulative_Savings_TRY'], label='Cumulative Savings', color='#2ca02c', linewidth=3)
plt.axhline(y=340155, color='#d62728', linestyle='--', label='Initial Investment (340,155 TRY)')
plt.axvline(x=31.1, color='orange', linestyle=':', label='Break-even Point (~31-38 Months)')

plt.fill_between(roi['Month'], roi['Cumulative_Savings_TRY'], 340155,
                 where=(roi['Cumulative_Savings_TRY'] > 340155), color='#e2f0e2', label='Net Profit Phase')

plt.xlabel('Months')
plt.ylabel('Total TRY')
plt.title('Return on Investment (ROI) Projection')
plt.legend()
plt.grid(True, alpha=0.3)
plt.savefig('roi_projection.png')

print("The analysis is complete. It is ready to be compiled into a report:)")
```
